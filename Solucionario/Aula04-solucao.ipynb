## <font color=#005757>1 Rato em um labirinto</font>

Considere um experimento fictício onde ratos recebem uma dose de medicamento e são colocados em um labirinto que eles devem completar. Os seguintes dados mostram a dose do medicamento administrada a um rato e o tempo que levou para o rato completar o labirinto.

| Dosagem (mg) | Tempo para completar o labirinto (s) |
|--------------|-------------------------------------|
| 0.0          | 105                                 |
| 0.08         | 98                                  |
| 0.2          | 54                                  |
| 0.37         | 50                                  |
| 0.6          | 65                                  |
| 0.84         | 81                                  |
| 1.02         | 182                                 |
| 1.2          | 210                                 |

Crie um gráfico de linha do tempo para completar o labirinto (no eixo y) versus dose (no eixo x). Faça a linha e o marcador substancialmente maiores e faça com que os intervalos dos eixos x e y comecem em zero e terminem no valor máximo para os respectivos eixos. Salve o gráfico com 300 dpi. Por favor, rotule os eixos apropriadamente.

***Extra:*** Escreva uma função que cria esse gráfico mas permite a passagem de valores de palavras-chave ao chamar a função para controlar a largura da linha, o tamanho do marcador e o nome do arquivo.

```python
import numpy as np
import matplotlib.pyplot as plt

dosagem = np.array([0.0, 0.08, 0.2, 0.37, 0.6, 0.84, 1.02, 1.2])
tempo = np.array([105, 98, 54, 50, 65, 81, 182, 210])

plt.plot(dosagem, tempo, "o--", lw = 2, ms = 10, c = "gray")
plt.axis([0, np.max(dosagem), 0, np.max(tempo)])

plt.xlabel("Dosagem (mg)")
plt.ylabel("Tempo para completar o labirinto (s)")

plt.savefig("rato_no_labirinto.png", dpi = 300)
```

```python
# Resposta do extra
def plotagem_rato_labirinto(x, y, largura_linha, tamanho_marker, nome_arquivo = "exercicio_1"):
  plt.plot(x, y, "o--", lw = largura_linha, ms = tamanho_marker, c = "gray")
  plt.axis([0, np.max(x), 0, np.max(y)])

  plt.xlabel("Dosagem (mg)")
  plt.ylabel("Tempo para completar o labirinto (s)")

  plt.savefig(f"{nome_arquivo}.png", dpi = 300)

plotagem_rato_labirinto(dosagem, tempo, 3, 20, "extra_rato_no_labirinto")
```

## <font color=#005757>1 Função Exponencial </font>

Crie uma tabela com cinco pontos de dados para valores de $x$ de zero a um e seus respectivos valores de $y$ para $y = e^x$. Crie um gráfico de linha dos valores de $y$ em relação aos valores de $x$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Criando valores de x:
x = [0, 0.2, 0.4, 0.6, 0.8, 1]

# Listas aceitam operações elemento a elemento:
y = np.exp(x)

# Plotando a função exponencial:
plt.plot(x, y)
plt.xlabel(f"Valores de x de {x[0]} até {x[-1]}")
plt.ylabel("Valores de y")
plt.show()
```

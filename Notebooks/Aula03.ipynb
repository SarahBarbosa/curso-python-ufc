# <font color=#005757>1. Exemplo de Criação de Gráficos de Linha e Dispersão</font>

Considere a seguinte lista fictícia de medições da quantidade de potência registrada por um sensor em vários momentos durante o curso de um dia:

| Hora decimal no dia | Potência (mW) |
|---------------------|---------------|
| 9.25                | 2.54          |
| 11                  | 4.10          |
| 13.5                | 1.21          |
| 15                  | 3.90          |
| 15.75               | 4.00          |

Uma representação decimal da hora 13:30 é 13.5. Vamos fazer um gráfico de linha e um gráfico de dispersão simples da potência versus hora decimal no dia:


````python
from matplotlib import pyplot as plt

hora_decimal = [9.25, 11, 13.5, 15, 15.75]
potencia = [2.54, 4.10, 1.21, 3.90, 4.00]

plt.plot(hora_decimal, potencia)
plt.xlabel("Hora decimal no dia")
plt.ylabel("Potência (mW)")
plt.show()
````

````python
plt.scatter(hora_decimal, potencia)
plt.xlabel("Hora decimal no dia")
plt.ylabel("Potência (mW)")
plt.show()
````

# <font color=#005757>2. Introdução a listas e tuplas </font>

## <font color=#005757>2.1 Listas </font>

Vimos anteriormente que as variáveis `hora_decimal`e `potencia` consistem em **listas** como entrada. Parece que uma lista é uma coleção de números, no entanto, elas podem conter praticamente qualquer coisa, mas por enquanto, não precisamos pensar sobre isso. Podemos descrever o básico das listas olhando para exemplos que consistem apenas de números.

Listas são sequências ordenadas. Elas são uma coleção de **elementos** onde o primeiro item tem a primeira posição, o segundo tem a segunda posição, e assim por diante. Cada uma dessas posições possui um *endereço* ou **índice** com o qual podemos nos referir ao item armazenado nessa posição.

Para nos referirmos a um elemento de uma lista que é atribuída a uma variável, usamos colchetes após o nome da variável, com o índice entre os colchetes:



````python
hora_decimal = [9.25, 11, 13.5, 15, 15.75]
hora_decimal
````

````python
hora_decimal[1]
````


Usando esta notação, podemos alterar os valores de uma lista existente ou usar os valores de uma lista existente:

Finalmente, o **comprimento** de uma lista (o número de elementos na lista) pode ser obtido usando a função embutida `len`:

````python
len(hora_decimal)
````

### <font color=#005757>2.2.1 Fatia de uma lista </font>

Ao especificar um intervalo de elementos para fatiar de uma lista, seguimos estas regras:

- Especificamos um intervalo de sublista usando valores de índice de elementos, fornecendo os limites inferior e superior do intervalo, separados por dois pontos.
- O limite inferior do intervalo é inclusivo, e o limite superior do intervalo é exclusivo.
- Se o limite inferior e/ou superior for omitido, a fatia vai até e através do início e/ou do final da lista, respectivamente.

````python
hora_decimal
````

````python
hora_decimal[0:-1]
````

````python
hora_decimal[-3:]
````

### <font color=#005757>2.2.2 Manipulação de listas </font>

As listas oferecem uma gama de funcionalidades por meio de **métodos**, que são funções associadas capazes de realizar operações específicas sobre ele. Os métodos de lista são cruciais para manipular seus elementos de forma eficaz e flexível:

- `append()`para adicionar elementos no final da lista:

````python
hora_decimal.append(20)
hora_decimal
````

- `pop()`para remover o último elemento da lista:

````python
hora_decimal.pop()
hora_decimal
````

- `insert()` para adicional um valor aonde você quiser via índice:

````python
hora_decimal.insert(1, 12)
hora_decimal
````

## <font color=#005757>2.3 Tuplas</font>

Nós vimos que uma lista é uma coleção ordenada de elementos cujos elementos podemos alterar. E se quisermos uma lista de itens que não queremos que seja alterada? Ou seja, queremos armazenar os itens e lê-los usando seus índices, mas não queremos poder substituí-los por valores diferentes. Para fazer isso, armazenamos a coleção de elementos como uma **tupla**. Definir uma tupla é igual a definir uma lista, exceto que usamos parênteses em torno dos elementos em vez de colchetes:

````python
hora_decimal = (9.25, 11, 13.5, 15, 15.75)
type(hora_decimal)
````

Finalmente, podemos converter listas em tuplas usando a função `tuple` e tuplas em listas usando a função `list`:

````python
hora_decimal = list(hora_decimal)
type(hora_decimal)
````

````python
hora_decimal = tuple(hora_decimal)
type(hora_decimal)
````

Uma maneira interessante para trabalhar com tuplas é usá-las para **descompactar valores**, vejamos um exemplo:

````python
def calcular_area_e_perimetro(retangulo):
    comprimento = retangulo[0]
    largura = retangulo[1]
    area = comprimento * largura
    perimetro = 2 * (comprimento + largura)
    return area, perimetro

retangulo = (5, 3)
area_retangulo, perimetro_retangulo = calcular_area_e_perimetro(retangulo)
area_retangulo, perimetro_retangulo
````

# <font color=#005757>3. Introdução a Strings </font>

Quando construímos os gráficos fizemos o seguinte:

```python
plt.xlabel("Hora decimal no dia")
plt.ylabel("Potência (mW)")
```

Cada uma dessas funções recebe um argumento, mas esse argumento não é um número (seja inteiro ou decimal) nem uma coleção de números (uma lista ou tupla), mas sim um novo tipo de variável chamado **string**. Strings são combinações de caracteres que são definidos entre apóstrofos correspondentes, aspas ou apóstrofos/três aspas/quadruplas aspas. Eles frequentemente representam informações de texto ou palavras, frases e sentenças.

````python
'Número de exoplanetas atuais'
"Número de exoplanetas atuais"
'''Número de exoplanetas atuais'''
"""Número de exoplanetas atuais"""
````

````python
type('Número de exoplanetas atuais')
````

Assim como podemos operar com números, também podemos operar com strings. Para concatenar duas strings, ou seja, uni-las, usamos o operador `+`:

````python
frase_1 = "Número de"
frase_2 = "atômos de carbono"
frase_1 + " " + frase_2
````

Além de unir duas strings, podemos extrair partes de strings, incluindo caracteres individuais. Para fazer isso, usamos uma sintaxe que é a mesma que a fatiação de listas e tuplas:

````python
texto = "Deus não joga dados"
texto[0:4]
````

Se quisermos saber o número de caracteres em uma string, podemos usar a função `len`:


````python
len(texto)
````

## <font color=#005757>3.1 F-strings </font>

Como mostrado na aula anterior, a função embutida do Python `print` desempenha um papel fundamental na exibição de informações, tal como mensagens e variáveis. No entanto, uma coisa interessante e importante para se observar sobre essa função são as [**f-strings**](https://docs.python.org/3/tutorial/inputoutput.html).

As f-strings fornecem uma maneira conveniente e eficiente de incorporar valores de variáveis dentro de strings, facilitando a formatação de saída. Com f-strings, podemos incluir variáveis diretamente dentro de uma string formatada, utilizando chaves `{}` precedidas por um `f` antes das aspas:

````python
numero_exoplanetas = 5_587
hoje = "3 de março de 2024"

print(f"Hoje, {hoje}, existem {numero_exoplanetas} exoplanetas descobertos.")
````

o caractere `_` é usado como um separador de dígitos em números inteiros. Esse recurso foi introduzido no Python 3.6 para melhorar a legibilidade de números grandes.

## <font color=#005757>3.2 Manipulação de strings </font>

Existem muitos métodos úteis para strings. Aqui nós examinamos alguns:

- `upper()`é usado para converter todos os caracteres de uma string em letras maiúsculas:

````python
titulo = "Uma introdução à programação Python - UFC"
titulo.upper()
````

- `lower()` é usado para converter todos os caracteres de uma string em letras minúsculas:

````python
titulo.lower()
````

- `title()` é usado para converter apenas as primeiras letras das palavras em maiúsculas:

````python
titulo.title()
````

- `split()` recebe uma string e retorna uma cópia dela como uma lista:

````python
titulo.split()
````

````python
titulo.split(sep = "-")
````

- `join()` pode ser pensado como o desfazendo do split:

````python
lista_do_titulo = ['Uma', 'introdução', 'à', 'programação', 'Python', '-', 'UFC']
" ".join(lista_do_titulo)
````

````python
print("\n".join(lista_do_titulo))
````

## <font color=#005757>4.1 Docstrings </font>

Outra forma de "comentar" é o uso de **docstrings**. *Docstrings descrevem funções*. O que os torna especiais é que, se seguirmos o formato padronizado para docstrings, a função de ajuda do Python automaticamente transforma o docstring em uma página de manual, com a formatação adequada:

````python
def porcentagem_para_decimal(entrada):
    """Converte uma porcentagem para seu equivalente decimal.

    O equivalente decimal é calculado dividindo a "entrada" por 100,
    e este valor é retornado pela função. O resultado será um float,
    independentemente de "entrada" ser float ou int.

    Assume-se que "entrada" é um escalar; nenhum teste é feito para
    verificar se isso é verdadeiro.

    Parâmetro de Entrada:
      entrada : float ou int

    Retorna:
      Decimal equivalente. Float.

    Exemplos:
    >>> print(porcentagem_para_decimal(42))
    0.42
    """
    return entrada / 100.0

porcentagem_para_decimal(42)
````

````python
help(porcentagem_para_decimal)
````

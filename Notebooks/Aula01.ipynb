# <font color=#005757>1. Exemplo de Python como uma Calculadora Básica </font>

Uma calculadora básica precisa fazer duas tarefas:
- Realizar operações aritméticas.
- Armazenar valores para reutilização.

Considere um cubo cujos lados têm 2 centímetros de comprimento. Aqui está o cálculo do volume deste cubo:

<div>
<img src="https://imgur.com/sFkxv5m.png" width="200"/>
</div>


```python
2 * 2 * 2
```

Podemos agrupar operações usando parênteses. Aqui está uma expressão que calcula o volume combinado de dois cubos, um com lados de 2 cm e outro com lados de 3 cm:

<div>
<img src="https://imgur.com/v6mVl1t.png" width="500"/>
</div>

```python
(2 * 2 * 2) + (3 * 3 * 3)
```

Finalmente, podemos salvar o resultado de uma expressão atribuindo um **variável**:

```python
volume_cubo_1 = 2 * 2 * 2
volume_cubo_2 = 3 * 3 * 3
soma_cubos = volume_cubo_1 + volume_cubo_2
soma_cubos
```

# <font color=#005757>2. Expressões </font>

**Expressões** são combinações de valores e variáveis manipuladas por **operadores**.


## <font color=#005757>2.1. Operadores </font>

**Valores** são números escritos como esperaríamos (por exemplo, 3, -7.2). Para *notação científica*, utilizamos `xey`, onde `x` é o número multiplicado pela potência de 10 e `y` é o expoente da potência de dez. Por exemplo, $5.4 \times 10^{6}$ é representado por:


```python
5.4e6
```

A tabela abaixo lista alguns **operadores aritméticos** básicos que o Python usa

Operação  |  Símbolo
---------|----------
Adição   |    +
Subtração|    -
Multiplicação| *
Divisão  |   /
Divisão Inteira| //
Exponenciação| **

Quando o Python avalia uma expressão aritmética, ele segue a ordem matemática padrão de blocos entre parênteses primeiro, depois exponenciação, depois multiplicação e divisão, e finalmente adição e subtração.

```python
4 + 3 * 2
```

```python
(4 + 3) * 2
```

O operador de divisão (`/`) atua como uma divisão regular, que retorna o quociente e o resto na forma decimal. Assim:







```python
1/2
```

```python
5/5
```

```python
5.0/5.0
```

Às vezes, no entanto, não queremos que a divisão nos dê o resultado decimal, mas sim apenas o quociente (como um número inteiro) e descartar o resto arredondando para baixo. Nestes casos, queremos usar a divisão inteira, que é dada pelo operador `//`.

```python
1//2
```

```python
5//5
```

## <font color=#005757>2.2 Variáveis </font>

Em Python, salvamos valores atribuindo-os a nomes de **variáveis** e usamos um sinal de igual para fazer a atribuição. Vamos chamar a variável que criamos no início desse notebook:

```python
volume_cubo_1
```

Além disso, Python também possui **operadores de atribuição combinados**, como `+=`, `-=`, `*=` e `/=`, que realizam uma operação aritmética e atribuem o resultado à variável. Por exemplo, isso

```python
volume_cubo_1 = volume_cubo_1 + 100
```

é o mesmo que isso,

```python
volume_cubo_1 += 100
```

## <font color=#005757>3.3 Raízes Usando a Fórmula Quadrática </font>

Considere o seguinte polinômio quadrático:

$$3x^2 + 4x - 3 = 0$$

Quais são as raízes de $x$? Resolva para $x$ usando a fórmula quadrática:

$$x = \dfrac{-b \pm \sqrt{b^2 - 4ac}}{2a} $$

onde $a$, $b$ e $c$ são os coeficientes do polinômio quadrático.
